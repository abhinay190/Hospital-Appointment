```python
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import plotly .express as px
```

```python
# import dataset
hs=pd.read_csv("hospital_appointment.csv")
hs.head(20)
```

```python
# checking shape
hs.shape
```

```python
# checking column names
hs.columns
```

```python
# checking info
hs.info()
```

```python
# checking describe
hs.describe().T
```

```python
# checking describe object
hs.describe(include="object")
```

```python
# checking duplicate rows
hs[hs.duplicated(keep=False)]
```

```python
# drop null values
hs.drop_duplicates(inplace=True)
```

```python
# checking null values
hs.isnull().sum()
```

```python
# rename columns
hs.rename(columns= {"Hipertension": "Hypertension", 
                     "Handcap":"Handicap"}, 
                                 inplace = True)
```

```python
hs.rename(columns= {"No-show": "Noshow", 
                     "SMS_received":"Smsreceived"}, 
                                 inplace = True)
```

```python
# check age columns
hs[hs["Age"]<0]
```

```python
# drop unexpected row
hs.drop(99832,axis=0,inplace=True)
```

```python
hs["Handicap"].unique()
```

```python
# replace handicap value 2,3,4 in 1
hs.replace({'Handicap': {2: 1, 3:1,4:1}},inplace=True)
```

```python
hs["Handicap"].unique()
```

```python
# drop columns PatientId,AppointmentID there is no need  in visualisation
hs.drop(['PatientId', 'AppointmentID'],axis=1,inplace=True)
```

```python
import datetime as dt
```

```python
# changing datetime object to datetime64[ns] 
hs['ScheduledDay'] = hs['ScheduledDay'].str[:10]
hs['AppointmentDay'] = hs['AppointmentDay'].str[:10]

# changing data type
hs['ScheduledDay'] = pd.to_datetime(hs['ScheduledDay'])
hs['AppointmentDay'] = pd.to_datetime(hs['AppointmentDay'])
```

```python
 hs[["ScheduledDay","AppointmentDay"]].dtypes
```

```python
# seprate numerical and categorical column using for loop
cat=[i for i in hs.columns if hs[i].dtypes=="object"]
num=[i for i in hs.columns if hs[i].dtypes!="object"]
```

```python
print(cat)
print(num)
```

```python
# checking unique values 
for i in hs.columns:
    if hs[i].dtypes in ["object","int64","float"]:
        print("Column Name=",i,"------","Unique_Values =",hs[i].unique())
```

```python
# extract dayofweek from scheduledday and appointmentday  
hs["Sche_week"]=hs["ScheduledDay"].dt.dayofweek
hs["Appo_week"]=hs["AppointmentDay"].dt.dayofweek
```

```python
hs.columns
```

```python
# drop ScheduledDay,AppointmentDay from columns
hs.drop(["ScheduledDay","AppointmentDay"],axis=1,inplace=True)
```

```python
sns.set_style("whitegrid") 
```

```python
# Gender distribution
ax = sns.countplot(x = 'Gender',data = hs)

for bars in ax.containers:
    ax.bar_label(bars)
plt.title("Gender distribution")
plt.show()
```

```python
# gender distribution with respect to noshow column 
ax=sns.countplot(x ="Gender", data = hs, hue = "Noshow")

for bars in ax.containers:
    ax.bar_label(bars)

plt.title("gender destribution with respect to noshow colunmn")
plt.show()
```

```python
# sms received destribution
ax=sns.countplot(x ="Smsreceived", data = hs)

for bars in ax.containers:
    ax.bar_label(bars)
plt.title("SMS received destribution")
plt.show()
```

```python
# sms destribution with respect to noshow column
ax=sns.countplot(x ="Smsreceived" , data = hs, hue ="Noshow")

for bars in ax.containers:
    ax.bar_label(bars)
plt.title('SMS destribution with respect to noshow colunmn')
plt.show()
```

```python
# scholarship v.s. no show with respect to gender
plt.figure(figsize=(20,5))
ax=sns.barplot(x = 'Scholarship', y = 'Noshow', hue = 'Gender', data = hs , errorbar=('ci',10))

for bars in ax.containers:
    ax.bar_label(bars)
plt.title('scholarship v.s. no show with respect to gender')

plt.show()
```

```python
# age bins
ax=hs['Age'].hist()
for bars in ax.containers:
    ax.bar_label(bars)
plt.show()
```
